```python
import pandas as pd
```

```python
df = pd.read_csv("data_complet_reorganiser.csv")
df
```

```python
# Supprimer doublons sur le couple (Cell line, variable)
df_unique = df.drop_duplicates(subset=["variable", "Cell line"])    #garder la première occurence de ("variable", "Cell line)

df_unique
```

```python
# Créer la table pivot
table = df_unique.pivot_table(
    index="variable",
    columns="Cell line",
    values="response",
    aggfunc="first"  # garde la première valeur si doublon
).fillna(100)  # remplacer NaN par 100

# Remettre l'ordre des colonnes si nécessaire
table = table.reset_index()
table.columns.name = None

# Limiter les valeurs max à 100
numeric_cols = table.select_dtypes(include="number").columns
table[numeric_cols] = table[numeric_cols].clip(upper=100)


# Liste de l'ordre souhaité
ordre = ["EFM192B", "KPL1", "MDAMB436", "OCUBM", "T-47D", "ZR751", "MDA-MB-468", "UACC-257", "BT-549", "HS 578T", "MCF7"]

# Réorganisation en gardant 'variable' en première colonne
table = table[["variable"] + ordre]
```

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import StandardScaler
```

```python
X_num = table.drop(columns=["variable"])
X = X_num.values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # X = tes données (77841, 11)

mbk = MiniBatchKMeans(n_clusters=4, batch_size=1000, random_state=42)
mbk.fit(X_scaled)

labels = mbk.labels_
```

```python
variables = table["variable"]
```

```python
# Créer un nouveau DataFrame avec 'variable' et les labels kmeans
result = pd.DataFrame({
    "variable": variables,
    "label": labels
})
```
